### Entrevista Carina Arvizu en Milenio.

Carina Arvizu Machado, aspirante a dirigir el Instituto de Planeación de la Ciudad de México presenta, de manera resumida, su visión de la planeación urbana en la Ciudad de México y su propuesta para el cargo. La entrevista se encuentra en la url: https://www.youtube.com/watch?v=KIqxHhQkWGI


```python
from youtube_transcript_api import YouTubeTranscriptApi
from google import genai
from fpdf import FPDF
```

### Extracción de texto de entrevista de Carina Arvizu, aspirante a dirigir el Instituto de Planeación de la CDMX

```python
#Leer credenciales de google

with open('C:/cred_google_gemini.txt', 'r') as file:
    llave = file.read()
```

```python
client=genai.Client(api_key=llave)
```

```python
srt = YouTubeTranscriptApi.get_transcript("KIqxHhQkWGI", languages=['es'])
```

```python
srt
```

```python
#Transcripción
entrevista=[]

#Extraer y pegar texto
for i in srt:
    entrevista.append(i['text'])
```

```python
#Convertir en string
entrevista_str=' '.join(entrevista)
entrevista_str
```

```python
#Salvar transcripción en txt
with open('entrevista_carina_arvizu.txt', 'w') as f:
    f.write(entrevista_str)
```

#### Resumen de la entrevista utilizando Google Gemini

```python

response = client.models.generate_content(
model="gemini-2.0-flash",
contents=[entrevista_str,"Genera una síntesis de lo planteado por Carina Arvizu en la entrevista de Milenio. Cuál es su visión de la planeación urbana en la Ciudad de México y cuál es su propuesta para el cargo de dirigir el Instituto de Planeación de la Ciudad de México. Su nombre es Carina"])

propuesta = response.text
propuesta
```

```python
#Guardar como pdf
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size = 12)
pdf.multi_cell(0, 10, propuesta)
pdf.output("resumen_carina_arvizu.pdf")
```
