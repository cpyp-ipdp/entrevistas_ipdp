```python
import whisper
import os
```

```python
# Cargar el modelo Whisper
model = whisper.load_model("base")
```

```python
#Enlistar archivos terminación .wav

archivos_mp3 = [f for f in os.listdir() if f.endswith('.mp3')]
```

```python
#Transcribir archivos y pegar en un solo elemento
transcripciones = []

for wav in archivos_mp3:
    ruta = os.path.join(os.getcwd(),
                        wav)
    transcripcion = model.transcribe(ruta,language="es")
    transcripciones.append(transcripcion)

transcripciones

```

```python
for item in transcripciones:
    print(type(item))

transcripciones = [str(item) if isinstance(item, dict) else item for item in transcripciones]
texto = " ".join(transcripciones)
```

```python
#Salvar texto como txt
with open("transcrip_entrevista_congreso.txt", "w", encoding="utf-8") as f:
    f.write(texto)
```
